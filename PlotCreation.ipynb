```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
textM = None
patternM = None
# Read the two CSV files generated in the Files folder.
try:
    textM = pd.read_csv("Files/RT_manipulating_text.csv")
except:
    print("Error Reading the file: Files/RT_manipulating_text.csv")
try:
    patternM = pd.read_csv("Files/RT_manipulating_pattern.csv")
except:
    print("Error Reading the file: Files/RT_manipulating_pattern.csv")
```

```python
# Show Dataframe textM
print(textM)

# Show Dataframe patternM
print(patternM)
```

```python
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the line plot
sns.lineplot(data=textM, x='n', y='KMPrt', ax=ax, label='KMPrt')
sns.lineplot(data=textM, x='n', y='RKrt', ax=ax, label='RKrt')
sns.lineplot(data=textM, x='n', y='LCSrt', ax=ax, label='LCSrt')
sns.lineplot(data=textM, x='n', y='O(n+m)', ax=ax, label='O(n+m)')

# Set the axis labels and title
ax.set_xlabel('n')
ax.set_ylabel('Runtime (in secs.)')
ax.set_title('Runtime: varying the text length [Non-magnified y-axis]')

# Show the plot
plt.show()
fig.savefig('Plots/plot_manipulating_text.png', format='png', dpi=1200, bbox_inches='tight')
```
